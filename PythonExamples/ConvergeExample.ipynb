## **convergence in distribution**

```python
from scipy.stats import norm
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Settings
n_vals = [5, 20, 100, 1000]
samples_per_n = 10000

# Set up the figure
#fig, axes = plt.subplots(1, len(n_vals), figsize=(18, 4), sharey=True)
x = np.linspace(-4, 4, 500)
standard_normal = 1/np.sqrt(2*np.pi) * np.exp(-x**2 / 2)

# Generate precise PDF curves for different Z_n and compare with Z ~ N(0,1)
fig, ax = plt.subplots(figsize=(10, 6))

# Plot standard normal distribution
ax.plot(x, standard_normal, 'r--', label='$Z \\sim \\mathcal{N}(0,1)$')

# Plot Z_n distributions
for n in n_vals:
    variance = 1 + 1/n
    pdf_zn = norm.pdf(x, loc=0, scale=np.sqrt(variance))
    ax.plot(x, pdf_zn, label=f'$Z_n \\sim \\mathcal{{N}}(0, 1 + 1/{n})$')

ax.set_title("Comparison of $Z_n$ Distributions with $Z \\sim \\mathcal{N}(0,1)$", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("Density")
ax.legend()
plt.grid(True)
plt.show()
```



---

### Distributions Plotted:

* **Red dashed line**: $Z \sim {N}(0, 1)$ — the limiting distribution.
* **Colored solid lines**: $Z_n \sim {N}(0, 1 + \frac{1}{n})$ for various values of $n$.

---

### What You See:

* As $n \to \infty$, the variance $1 + \frac{1}{n} \to 1$, and the curves approach the standard normal.
* For small $n$, the curves are slightly wider (larger variance).
* All $Z_n$ are centered at 0, but with shrinking variance.

---
### Conclusion:

This confirms that $Z_n \xrightarrow{D} {N}(0,1)$, i.e., **convergence in distribution**, but not necessarily convergence in probability or almost surely — since for each $n$, $Z_n$ is still random with positive variance.
